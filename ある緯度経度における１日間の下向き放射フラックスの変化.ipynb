```python
import os
import subprocess
import pygrib

# 2020/08/01 15:00(UTC 06:00)の地上データをダウンロードする
cwd = os.getcwd()
url_surf = 'http://database.rish.kyoto-u.ac.jp/arch/jmadata/data/gpv/original/2018/10/06/Z__C_RJTD_20181006120000_GSM_GPV_Rjp_Lsurf_FD0000-0312_grib2.bin'
subprocess.run(['curl', '-O', url_surf], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, cwd=cwd)

## GRIB2ファイルを読み込む
file_surf = os.path.join(cwd, os.path.basename(url_surf))
grbs = pygrib.open(file_surf)
```

```python
for grb in grbs:
  print(grb)
```

```python
import numpy as np
data = np.zeros(24)
for i in range(24):
  grb = grbs.select(parameterName='Downward short-wave radiation flux',forecastTime=i+3)[0]
  values, lats, lons = grb.data(lat1=35.5, lat2=35.6, lon1=139.6, lon2=139.75)
  data[i] = values[0][0]
print(data)
```

```python
import numpy as np
np.save(
    "20181007_rad_time",  # ファイル名
    data # 保存したいオブジェクト
)
```
